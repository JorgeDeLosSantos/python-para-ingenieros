## La recta tangente

La curva $y = 1/(1+x^2)$ se llama **bruja de María Agnesi**. Encuentre la ecuación de la recta tangente a esta curva en el punto $(-1,1/2)$.

La recta tangente de una curva plana cualesquiera en el punto $ (a, f(a)) $ es aquella de pendiente $f'(a)$ y que pasa por dicho punto. De la forma punto-pendiente de una recta se sabe que la ecuación de la recta tangente estará dada por:

$$ y = f'(a) \left( x - a \right) + f(a) $$

Primeramente, definimos la variable simbólica $x$, la función $y$ y el punto por el cual pasa la recta:

```python
px,py = -1, 1/2
x = symbols("x")
y = 1/(1+x**2)
```

Calculamos la derivada de la función:

```python
dy = diff(y, x)
```

La pendiente de la recta tangente será entonces lo que resulte de evaluar la derivada en la coordenada correspondiente:

```python
m = yp.subs({x:px})
```

Entonces la ecuación de la recta tangente estará dada por:

```python
m*(x - px) + py
```

## Y rectas tangentes ...

```python
x = symbols("x")
y = 3*sin(x/5)
yp = diff(y, x)
s = np.linspace(-10,10)

for px in np.linspace(-5,5,10):
    m = yp.subs({x:px})
    py = y.subs({x:px})
    rt = m*(x - px) + py
    nrt = lambdify(x, rt, "numpy")
    plt.plot(s,nrt(s),"--")
    
plt.plot(s, lambdify(x, y, "numpy")(s), "r", lw=2);
```
